# 快速指南：如何使用练习文件

使用 Poetry 进行依赖管理的 GitHub 存储库的工作包括几个简单的步骤。以下是如何下载存储库、设置并开始使用 Poetry 和 Poetry shell 进行工作的简单指南。

#### 第一步：下载或克隆 GitHub 存储库

首先，您需要将文件下载到您的计算机。您可以通过 GitHub 网站轻松完成此操作。以下是将存储库作为 ZIP 文件下载的分步骤指南：

* 使用任何网络浏览器导航到 GitHub 上存储库的页面。
* 找到位于存储库页面顶部附近的绿色按钮 **代码**。
* 点击此按钮，您将看到一个选项 **下载 ZIP**。
* 点击 **下载 ZIP**。这将把整个存储库的 ZIP 文件下载到您的计算机上。
* 在您的下载文件夹或通常保存下载的地方找到 ZIP 文件。将 ZIP 文件中的文件提取到您选择的计算机目录。通常可以通过右键点击 ZIP 文件并选择一个选项，比如“提取全部”、“解压缩”或“在此提取”，具体取决于您的操作系统。

我们建议您按照前面步骤下载该存储库。如果您更喜欢从 GitHub 克隆存储库到本地机器，您需要安装 Git，并在终端中输入以下内容：

**示例**：
```bash
git clone https://github.com/username/repository-name.git
```
将 `https://github.com/username/repository-name.git` 替换为实际的 GitHub 存储库 URL。

#### 第二步：导航到项目目录

下载完成后，您需要导航到项目目录。

**示例**：
```bash
cd project-name
```
将 `project-name` 替换为 Git 下载或克隆存储库时创建的文件夹名称。

#### 第三步：确保您在目录中激活了正确的 Python 版本
```bash
python --version
```
如果您没有激活 Python 3.11.4，请确保您在计算机上安装了 [Python 3.11.4](https://www.python.org/downloads/) 并通过在终端中执行以下命令将其激活：
```bash
pyenv local 3.11.4
```

#### 第四步：使用 Poetry 安装依赖

在您开始工作之前，您需要安装项目的 `pyproject.toml` 文件中指定的所有依赖。如果项目是使用 Poetry 设置的，您应该在项目目录中看到此文件。

**示例**：
```bash
poetry install
```
此命令告知 Poetry 读取 `pyproject.toml` 和 `poetry.lock` 文件，并将所有必要的依赖项安装到专用于此项目的虚拟环境中。

#### 第五步：激活 Poetry Shell

要在 Poetry 管理的虚拟环境中工作，您可以激活 Poetry shell。这将确保您安装的任何 Python 命令、脚本或额外软件包都限制在此环境中，从而防止与其他项目或系统范围的 Python 软件包发生冲突。

**示例**：
```bash
poetry shell
```
执行此命令后，您的终端现在将在虚拟环境中运行。您通常会看到环境名称前缀出现在终端提示符中。

#### 第六步：打开项目中包含的 Jupyter Notebooks

项目中已经安装了 Jupyter Lab。为了在浏览器中打开它并打开包含在项目中的 notebooks，请在终端中运行以下命令。

**示例**：
```bash
jupyter lab
```
执行此命令后，jupyter lab 将在您的浏览器中打开。前往那里，打开文件目录并选择您想要打开的 Jupyter notebook。
这些步骤提供了一种干净且隔离的方式来处理使用 Poetry 的 Python 项目，确保所有依赖项得到正确管理，冲突最小化。这种方法在贡献或测试开发环境中的项目时特别有用。
